#### Buscar dados da Ação automaticamente

Bibliotecas: `yfinance` e `matplotlib`

`pip install yfinance`
`pip install matplotlib`

```python
# importando a yfinance
import yfinance 
```

```python
dados = yfinance.Ticker("PETR4.SA").history(start="2020-01-01", end="2020-12-31")
fechamento = dados.Close
#dados.Open
fechamento.plot()
```

```python
dados = yfinance.Ticker("BBAS3.SA").history(start="2020-01-01", end="2020-12-31")
fechamento = dados.Close
fechamento.plot()
```

```python
# Aqui vc adciona o Codigo da Ação para a analise (Ex. BBAS3.SA, PETR4.SA)
ticker = input("Digite o codigo da ação desejada: ")

# Aqui vc Digita a Data inicial
inicio = input("Digite a Data Inicial: (Ano) XXXX- (Mes) XX- (Dia) XX")

# Aqui vc Digita a Data Final
fim = input('Digite a Data Final:(Ano) XXXX- (Mes) XX- (Dia) XX')

#dados = yfinance.Ticker(ticker).history(start="2020-12-01", end="2021-01-01")
dados = yfinance.Ticker(ticker).history(start=inicio, end=fim)
fechamento = dados.Close
fechamento.plot()

print("Código digitado foi: ", ticker)
print("Data do Inicio: ", inicio)
print("Data do Fim: ", fim)
```

#### Criar as análise solicitadas

- Cotação máxima
- Cotação mínima
- Valor médio

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
valor_medio = round(fechamento.mean(), 2)

print(maxima)
print(minima)
print(valor_medio)

```

O `Round()` Organiza e formata a apresentação dos numeros apresentados no caso

#### Enviar o e-mail automaticamente

- abrir o navegador e ir ao Gmail
- clicar no botao Escrever
- Digitar o email do destinatário e teclar TAB
- Digitar o assunto e teclar TAB
- Digitar a mensagem
- Clicar no botão Enviar

Bibliotecas: `pyautogui`, `pyperclip`, `webbrowser`

`pip install pyautogui`


```python
import pyautogui
import pyperclip
import webbrowser
import time

destinatario1 = "mukito100@gmail.com"
# destinatario2 = "mukitoprograma@gmail.com"
assunto = "Análises do Projeto 2020"
mensagem = f"""

Prezado gestor,

Segue as análises solicitadas da ação {ticker}:

Cotação máxima: R${maxima}
Cotação Mínima: R${minima}
Valor médio: R${valor_medio}

Qualquer dúvida, estou à disposição!

Atenciosamente
Fabiano Santos Ramos
Analista de Sistemas
(61)98646-2326

"""
#, ticker, maxima, minima, valor_medio
# abrir o navegador e ir para o gmail
webbrowser.open("www.gmail.com")
time.sleep(3)

# configurando uma pausa de 3 segundos
pyautogui.PAUSE = 3

# clicar no botao escrever
pyautogui.click(x=-1834, y=189)

# digitar o email do destinatario e teclar TAB
pyperclip.copy(destinatario1)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digitar o assunto do email
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digitar a mensagem
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# clicando para enviar
pyautogui.click(x=-602, y=1006)
```
